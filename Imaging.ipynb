```python
import pandas as pd
```

```python
results_df = pd.read_csv('1345_customer_results.csv',
                         usecols = ['Origin', 'Origin Created At', 'Labeling State', 'Qualified Reads', 'Correct Label', 
                                    'Majority Label', 'Difficulty', 'Agreement', 'First Choice Answer', 'First Choice Votes', 
                                    'First Choice Weight', 'Second Choice Answer', 'Second Choice Votes', 'Second Choice Weight'])
pd.set_option('display.max_colwidth', None)

results_df
```

- Removed the extra empty columns that won't serve as anything useful

```python
results_df['Expert Yes Votes'] = results_df['Origin'].str[-5]
results_df['Expert Yes Votes'] = results_df['Expert Yes Votes'].astype(int)
```

```python
results_df['Expert Yes Votes'].value_counts()
```

```python
#Drop rows where there are 9 votes:
results_df = results_df[results_df['Expert Yes Votes'] < 9]
```

```python
results_df['Expert Yes Votes'].value_counts()
```

```python
#Expert majority answers:
results_df.loc[results_df['Expert Yes Votes'] > 4,'Expert Label'] = "'yes'"
results_df.loc[results_df['Expert Yes Votes'] < 4,'Expert Label'] = "'no'"
```

```python
results_df
```

# Does the expert majority agree with the crowd majority?

```python
results_df.loc[results_df["Expert Label"] == results_df["Majority Label"],'Expert Label Agrees w/ Crowd Majority?'] = "'agree'"
results_df.loc[results_df["Expert Label"] != results_df["Majority Label"],'Expert Label Agrees w/ Crowd Majority?'] = "'disagree'"
```

```python
results_df["Expert Label Agrees w/ Crowd Majority?"].value_counts('agree')
```

- In 70.6661% of the 30279 counts of data, the expert majority agrees with the crowd majority

# Who Performs Better: Crowd or Experts?

```python
results_df.loc[results_df["Expert Label"] == results_df["Correct Label"],'Expert Label Correct?'] = "'correct'"
results_df.loc[results_df["Expert Label"] != results_df["Correct Label"],'Expert Label Correct?'] = "'incorrect'"

results_df.loc[results_df["Majority Label"] == results_df["Correct Label"],'Majority Label Correct?'] = "'correct'"
results_df.loc[results_df["Majority Label"] != results_df["Correct Label"],'Majority Label Correct?'] = "'incorrect'"
```

```python
results_df
```

```python
results_df['Expert Label Correct?'].value_counts()
```

```python
results_df['Majority Label Correct?'].value_counts()
```

```python
#Expert Correct Percentage:
results_df['Expert Label Correct?'].value_counts('correct')
```

```python
#Majority Correct Percentage:
results_df['Majority Label Correct?'].value_counts('correct')
```
